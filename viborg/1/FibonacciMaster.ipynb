```python
import time
from tqdm import tqdm 
import plotly
from plotly.offline import iplot, init_notebook_mode
from IPython.display import display, HTML
init_notebook_mode(connected=True)
from fibplot import fibplot
```

```python
# define global variables
count = 0
```

```python
# Recursive Fibonacci Function
def fibr(n):
    global count
    count += 1
    if n == 1 or n == 2:
        return 1
    else:
        return fibr(n-1) + fibr(n-2)
```

```python
# Imperative Fibonacci Function
def fibi(n):
    global count
    fibs = [1,1]
    while(len(fibs) < n):
        count+=1
        fibs.append(fibs[-1] + fibs[-2])
    return fibs[-1]
```

```python
# Run recursive fibonacci function from i = 1..20
plotvalsr = []
for i in range(20):
    count = 0
    fibr(i+1)
    plotvalsr.append([i+1,count])
```

```python
# Run imperative fibonacci function from i = 1..20
plotvalsi = []
for i in range(20):
    count = 0
    fibi(i+1)
    plotvalsi.append([i+1,count])
```

```python
# Extract x and y values for each fibonacci function, to be plotted
xsr,ysr = [x for [x,y] in plotvalsr], [y for [x,y] in plotvalsr]
xsi,ysi = [x for [x,y] in plotvalsi],[y for [x,y] in plotvalsi]
```

```python
# p-p-p-p-p-ploooooot
fibplot(xsr,ysr,xsi,ysi)
```
